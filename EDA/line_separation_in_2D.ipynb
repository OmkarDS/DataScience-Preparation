```python
filepath = "/content/drive/MyDrive/Data/iris_csv.csv"
iris = pd.read_csv(filepath)

sns.set_style("whitegrid")
facetgrid = sns.FacetGrid(data = iris, hue="class", height=5)
facetgrid.map(plt.scatter, 'petallength', 'petalwidth')

plt.plot(x, -0.25*x+1.5, linestyle='solid')

facetgrid.add_legend()

plt.show()
```

```python
iris_setosa = iris.loc[iris['class'] == 'Iris-setosa']
other = iris.loc[iris['class'] != 'Iris-setosa']
```

```python
other.head()
```

```python
setosa_petal_lengths = iris_setosa['petallength'].reset_index(drop=True)
setosa_petal_widths = iris_setosa['petalwidth'].reset_index(drop=True)
other_petal_lengths = other['petallength'].reset_index(drop=True)
other_petal_widths = other['petalwidth'].reset_index(drop=True)
```

```python
other_petal_lengths
```

```python
def line_equation_value(x, y):
    return y + 0.25*x - 1.5
```

```python
setosa_values = [line_equation_value(setosa_petal_lengths[i], setosa_petal_widths[i]) for i in range(len(setosa_petal_widths))]
print(f"Number of setosa points that are in opposite direction of normal to the line : {len(list(filter(lambda x : x>0, setosa_values)))}")
```

```python
other_values = [line_equation_value(other_petal_lengths[i], other_petal_widths[i]) for i in range(len(other_petal_widths))]
print(f"Number of versicolor and virginica points that are in same direction of normal to the line : {len(list(filter(lambda x : x>0, other_values)))}")
```
